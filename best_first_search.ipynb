```python
# connessioni tra cittÃ 

connections = {}

connections['A'] = {(3, 'B'), (4, 'C'), (5, 'E')}
connections['B'] = {(2, 'D'), (4, 'A'), (7, 'E'), (3, 'G')}
connections['C'] = {(7, 'F'), (1, 'A'), (10, 'H')}
connections['D'] = {(3, 'B'), (4, 'C')}
connections['E'] = {(5, 'C')}
connections['F'] = {(2, 'H'), (1, 'E')}
connections['G'] = {(6, 'C')}
connections['H'] = {(3, 'C'), (3, 'B')}
```

```python
class Node:
    
    def __init__(self, state, parent, cost):
        self.state = state
        self.depth = 0
        self.children = []
        self.parent = parent
        self.cost = cost
        
        
    def addChild(self, childNode):
        self.children.append(childNode)
        childNode.parent = self
        childNode.depth = self.depth + 1
        
    
    def printPath(self):
        if self.parent != None:
            self.parent.printPath()
        print("-> ", self.state.name)
    
    def __lt__(self, other):
        return 1
```

```python
class State:
    
    def __init__(self, name = None):
        if name == None:
            self.name = self.getInitialState()   # crea lo stato iniziale
        else:
            self.name = name
            
    def getInitialState(state):
        initialState = 'A'
        return initialState
    
    def successorFunction(self):
        return connections[self.name]
    
    def checkGoalState(self):
        return self.name == 'H'
```

```python
import queue

def best_first_search():

    fringe = queue.PriorityQueue()
    visited = []
    
    initialState = State()
    
    root = Node(initialState, None, 0)
       
    fringe.put((root.cost, root))
    
    while not fringe.empty():

        _, currentNode = fringe.get()
        
        print("-- dequeue --", currentNode.state.name, "a costo", currentNode.cost)
        
        visited.append((currentNode.cost, currentNode.state.name))
        
        if currentNode.state.checkGoalState():
            print("Stato obiettivo raggiunto a costo", currentNode.cost)          
            print("----------------------")
            print("Soluzione:")
            currentNode.printPath()
            break
            
        for (childCost, childState) in currentNode.state.successorFunction():
            childNode = Node(State(childState), currentNode, childCost + currentNode.cost)  
            
            if (childNode.cost, childNode.state.name) not in visited:
                
                currentNode.addChild(childNode)
                fringe.put((childNode.cost, childNode))
```

```python
best_first_search()
```
